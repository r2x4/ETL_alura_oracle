```python

# Leer el JSON

import pandas as pd

datos = pd.read_json('datos_hosting.json')
```

```python
datos.head()
```

# **Normalizar los datos JSON**

```python
# normalizar datos

datos = pd.json_normalize(datos['info_inmuebles'])
```

```python
datos.head()
```

# **Extraer listas de listas o listas de diccionarios**

```python
# extraer listas

columnas = list(datos.columns)
columnas
```

```python
# explotar datos sesaca lista d elistas o lista en dicionarios en JSON

datos = datos.explode(columnas[3:])
datos
```

# Receteamos los indices por que no pueden ser cero y quedan por explotar los diccionarios

```python
# se reinician los indices y se borran y sobrescriben en la memoria

datos.reset_index(inplace=True, drop=True)
datos.head()
```

```python
# datos aleatorios

datos.sample(7)
```

```python
# informacion de los datos

datos.info()
```

# **Datos Numericos**

```python
import numpy as np
```

```python
# se cambia el tipo de datos

datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)

# funciona solo si son todos enteros ojo
```

```python
datos.info()
```

```python
# creamos lista para columnas numericas

col_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
datos[col_numericas] = datos[col_numericas].astype(np.int64)
```

```python
datos.info()
```

# **Datos Float**

```python
# float

datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
datos.info()
```

# **Hay float con dolar ocomas se tartan asi**

```python
# reemplaza el $ por nada el segundo replace es ejemplo

datos['precio'] = datos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
# ahora si cambiamos el dato

datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

# **Varias columnas**

```python
# dos corchetes de dos columnas hacia arriba se usa applymap

datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
# ahora si cambiamos el dato

datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

# **Datos de texto o strings**

### haremos tocanizar o hacer tokens de palabras

```python
# haceremos tokens ver por series lower para minisculas todo

datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
datos.head()
```

```python
datos['descripcion_local'][3169]
```

```python
# se quitan caracteres especiales

datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
# se quitan caracteres especiales

datos['descripcion_local'] = datos['descripcion_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
# para los espacios en blanco creados con los dos regex

datos['descripcion_local'].head()
```

```python
# separacion de las palabras con comas

datos['descripcion_local'] = datos['descripcion_local'].str.split()
```

```python
datos.head()
```

```python
# columnas con diccionarios en listas

datos['comodidades'] = datos['comodidades'].str.replace('\{|}|\"', '', regex=True)
```
