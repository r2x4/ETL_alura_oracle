## **Normalizar el JSON**

### Usar Visualizador JSON CRACK se ve el diagrama de diseno

```python
datos_normalizar
```

# **Exportando o escribiendo un JSON**

```python
# Exportar

datos_normalizar.to_json('historico_pacientes_norm.json')
```

# Ahora a de JSON a csv

```python
# Exportar a CSV

datos_normalizar.to_csv('historico_pacientes_norm.csv', index=False)
```

# **Practica**

### Leer datos de un API tipo JSON

```python
import requests
import pandas as pd
from pandas import json_normalize

url = 'https://www.fruityvice.com/api/fruit/all'
datos = requests.get(url).json()

datos_api = json_normalize(datos)



```

```python
datos_api.head()
```

```python
# exportar a JSON

datos_api.to_json('datos_api.json')
```
