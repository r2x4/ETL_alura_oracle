# **Demandas de ML**

### **¿Cuáles son los valores promedio de alquiler por tipo de propiedad?**

```python
# ML = Machine_Learning

datos.head()
```

```python
# promedio de los alquileres global

datos['Valor'].mean()
```

```python
# agrupar datos

datos.groupby('Tipo').mean(numeric_only=True)
```

```python
# agrupar datos por columna y calcular promedio de dicha columna

datos.groupby('Tipo')['Valor'].mean()
```

```python
# los organizamos tipo dataframe y evaluamos de menor a mayor

datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
# lo almacenamos en una variable y graficarlo

df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh', figsize=(12,8), color='purple')
```

```python
# valores unicos

datos.Tipo.unique()
```

```python
# inmubles comerciales

inmuebles_comerciales = ['Conjunto Comercial/Sala', 'Edificio Completo', 'Tienda/Salón', 'Casa Comercial',
                         'Terreno Estándar', 'Cochera/Estacionamiento', 'Galpón/Depósito/Almacén',
                         'Tienda en Centro Comercial','Hotel', 'Loteo/Condominio','Industria']
```

```python
# inmuebles comerciales

datos.query('@inmuebles_comerciales in Tipo')
```

```python
# inmuebles recidenciales

datos.query('@inmuebles_comerciales not in Tipo')
```

```python
# hacer el dataframe

df = datos.query('@inmuebles_comerciales not in Tipo')
df.head()
```

```python
# lo almacenamos en una variable y graficarlo

df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh', figsize=(12,8), color='purple')
```

```python
df.Tipo.unique()
```

```python
# frecuencia

df.Tipo.value_counts()
```

```python
# porcentajes por tipos

df.Tipo.value_counts(normalize=True)
```

```python
# graficar pocentajes

porcentaje_propiedades = df.Tipo.value_counts(normalize=True).mul(100).to_frame()
porcentaje_propiedades.plot(kind='bar', figsize=(12,8), color='green',xlabel='Tipo',ylabel='Porcentaje')
```

```python
# equipo de ML solicita el trabajar con el mayor

df.query('Tipo == "Departamento"')
```

```python
# lo guardamos en un adatframe

df_departamento = df.query('Tipo == "Departamento"')
df_departamento.head()
```
