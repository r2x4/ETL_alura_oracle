### **y = ax + b**

```python
# el 1.5 es un valor de crecimiento del precio de las manzans por mes es hipotetico

x = fechas
y = 1.5 * x + 78
```

```python
y
```

```python
# potenciar para buscar resultados 

np.sum(np.power(Moscu - y,2))
```

```python
# luego sacra raiz cuadrada 

np.sqrt(np.sum(np.power(Moscu - y,2)))
```

```python
# forma facil jajaja

np.linalg.norm(Moscu - y)
```
