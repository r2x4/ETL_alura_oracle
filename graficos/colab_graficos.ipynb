# **Iniciamos el tratamiento de Datos**

```python
# tomamos un indice osea la primera columna

df.set_index('Pais', inplace=True)
df.head()

```

# **Creamos un dataframe y una variable para ver colombia**

```python
# creamos la variable


años = list(map(str, range(1980, 2014)))
años
```

```python
# variable colombia

colombia = df.loc['Colombia', años]
colombia
```

```python
# hacer diccionario

col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes': colombia.values.tolist()}
```

# **Creamos el DataFrame Colombia**

```python
# creamos el DataFrame Colombia

datos_col = pd.DataFrame(col_dict)
datos_col.head()
```

# **Matplotlib**

https://matplotlib.org/

```python
# graficamos

import matplotlib.pyplot as plt
```

```python
# graficamos

plt.figure(figsize=(8, 4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014'])
plt.title('Inmigrantes en Colombia hacia Canada')
plt.xlabel('Año')
plt.ylabel('Numero de Inmigrantes')
plt.show()
```

# **Almacenar en Variables**

```python
# almacenar variable

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes en Colombia hacia Canada\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('Numero de Inmigrantes')
plt.show()
```

# **Grafico de Caja**

```python
# crear lienzo y variables

fig, axs = plt.subplots(1,2,figsize=(10, 3))
axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].set_title('Inmigrantes en Colombia hacia Canada\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('Numero de Inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid(True)


axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Bloxplot de la inmigracion de colombianos\nhacia Canada 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('Numero de Inmigrantes')
axs[1].grid(True)
plt.show()
```

```python
# corroborar

datos_col.describe()
```

# **Otros Paises**

```python
# varios graficos

fig, axs = plt.subplots(2,2,figsize=(10, 6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)

axs[0,0].plot(df.loc['Brasil', años])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colombia', años])
axs[0,1].set_title('Colombia')

axs[1,0].plot(df.loc['Perú', años])
axs[1,0].set_title('Perú')

axs[1,1].plot(df.loc['Argentina', años])
axs[1,1].set_title('Argentina')

# ajustarlas todas

for ax in axs.flat:
  ax.set(xlabel='Año', ylabel='Numero de Inmigrantes')
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid(True)

plt.show()
```

```python
# almacenar variable

fig, ax = plt.subplots(figsize=(8, 3))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'],lw=2,color='g')#,marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes en Colombia hacia Canada\n1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('Año',fontsize=14)
ax.set_ylabel('Numero de Inmigrantes',fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
plt.show()
```

# **Haremos otro DataFrame America del Sur**

```python
# dataframe

sudamerica = df.query('Region == "América del Sur"')
sudamerica.head(10)
```

```python
# lista de 10 colores

colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b',
           '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#aec7e8', '#ffbb78']

fig, ax = plt.subplots(figsize=(12, 5))

ax.bar(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo 1980 - 2013',
             loc='left',fontsize=18)
ax.set_xlabel('Pais')
ax.set_ylabel('Numero de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12,rotation=90)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
# lista de 10 colores barras horizontales

colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b',
           '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#aec7e8', '#ffbb78']

fig, ax = plt.subplots(figsize=(12, 5))

ax.barh(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo 1980 - 2013',
             loc='left',fontsize=18)
ax.set_ylabel('Pais')
ax.set_xlabel('Numero de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values(by='Total', ascending=True)

colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b',
           '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#aec7e8', '#ffbb78']

fig, ax = plt.subplots(figsize=(12, 5))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigracion de sudamericanos hacia Canada\ndurante el periodo 1980 - 2013',
             loc='left',fontsize=18)
ax.set_ylabel('Pais')
ax.set_xlabel('Numero de Inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
# enfoque a colombia

colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('silver')

sudamerica_sorted = sudamerica.sort_values(by='Total', ascending=True)


fig, ax = plt.subplots(figsize=(12, 5))

ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='y', which='both', length=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudamerica : Colombia es el segundo pais con mas migrantes\nhacia Canada durante el periodo 1980 - 2013',
             loc='left',fontsize=18)
ax.set_ylabel('Pais')
ax.yaxis.set_tick_params(labelsize=12)
#ax.set_xlabel('Numero de Inmigrantes', fontsize=14)
#ax.xaxis.set_tick_params(labelsize=12)

for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20, i, str(j),color='Black',fontsize=8.5, ha='left', va='center')

plt.show()
```

# **Graficos en Seaborn**

https://seaborn.pydata.org/

```python
# seaborn

import seaborn as sns
```

```python
# Dejar un tema inicial
sns.set_theme()
```

```python
# tpo 10

top_10 = df.sort_values(by='Total', ascending=False).head(10)
top_10
```

```python
# hacer grafico

ax = sns.barplot(data=top_10,x='Total', y=top_10.index,orient='h')
ax.set(title='Top 10 de paises con mayor migracion hacia Canada\n1980 -2013',
        xlabel='Numero de Inmigrantes',
        ylabel='Año')

plt.show()
```

```python
# crear el linezo o figura

fig, ax = plt.subplots(figsize=(10, 5))

ax = sns.barplot(data=top_10,x='Total', y=top_10.index,orient='h',palette='coolwarm',
                 hue=top_10.index, legend=False)
ax.set_title('Top 10 de paises con mayor migracion hacia Canada\n1980 -2013',
        loc='left',fontsize=16)
ax.set_xlabel('Numero de Inmigrantes',fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
# funcion para paleta de colores

def generar_graficos(palette):
  fig, ax = plt.subplots(figsize=(10, 5))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y', which='both', length=0)

  ax = sns.barplot(data=top_10,x='Total', y=top_10.index,orient='h',palette=palette,
                 hue=top_10.index, legend=False)
  ax.set_title('Top 10 de paises con mayor migracion hacia Canada\n1980 -2013',
        loc='left',fontsize=16)
  ax.set_xlabel('Numero de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  ax.xaxis.set_tick_params(labelsize=12)
  ax.yaxis.set_tick_params(labelsize=12)

  for i,j in enumerate(top_10['Total']):
    ax.text(j+20, i, str(j),color='Black',fontsize=8.5, ha='left', va='center')


  plt.show()
```

```python
generar_graficos('Spectral')
```

```python
# mejor opcion una categorica

generar_graficos('Paired')
```

```python
# funcion para paleta de colores

def generar_graficos1(palette):
  fig, ax = plt.subplots(figsize=(10, 5))
  ax.set(xticklabels=[])
  ax = sns.barplot(data=top_10,x='Total', y=top_10.index,orient='h',palette=palette,
                 hue=top_10.index, legend=False)
  ax.set_title('Top 10 de paises con mayor migracion hacia Canada\n1980 -2013',
        loc='left',fontsize=16)
  ax.set_xlabel('Numero de Inmigrantes',fontsize=14)
  ax.set_ylabel('')
  ax.xaxis.set_tick_params(labelsize=12)
  ax.yaxis.set_tick_params(labelsize=12)
  sns.despine()

  for i,j in enumerate(top_10['Total']):
    ax.text(j+20, i, str(j),color='Black',fontsize=8.5, ha='left', va='center')


  plt.show()
```

# **Plotly**

https://plotly.com/python/

```python
# importamos plotly

import plotly.express as px
```

```python
# graficar

fig = px.line(datos_col,x='Año',y='Inmigrantes',
            title='Inmigracion de Colombianos hacia Canada<br>1980 - 2013')
fig.update_traces(line_color='green',line_width=3)
fig.update_layout(width=800,height=400, xaxis={'tickangle':-45},
                  xaxis_title='Año',yaxis_title='Numero de Inmigrantes',
                  font_family='firacode',font_size=14, font_color='grey')
fig.show()
```
